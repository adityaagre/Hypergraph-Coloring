```python
Proper_coloring(hypergraph_object):
    # Function to do complete proper coloring for a given hypergraph
    
    if(hypergraph_object.ongoing_color == False):
        ## If we have ongoing color as false, the inference is that we are no longer looking for vertices 
        ## compatible with a previous vertex. The last color has een used as often as it can and we must hence 
        ## assign a new color for any forthcoming uncolored vertices.
        
        for vertex in hypergraph_object.vertices:
            ## Iterating through all hypergraph vertices.
            
            if(vertex is uncolored):
                ## We have atleast one uncolored vertex.
                ## We need a new color
                hypergraph_object.colors_used++
                hypergraph_object.current_color = colors_used
                
                ## Since this color is new, no vertices in the hypergraph have this color
                ## So, reset the current_color_vertices list in the hypergraph
                hypergraph_object.current_color_vertices = []
                
                ## Similarly, reset the repeatable list too
                hypergraph_object.repeatable_list_clear()
                
                
                ## Assign this new color to vertex in consideration
                hypergraph_object.assign_color(vertex,current_color)
                hypergraph_object.latest_colored_vertex = vertex
                
                ## Add this vertex to current_color_vertices of hypergraph
                hypergraph_object.current_color_vertices.append(vertex)
                
                ## Now, set ongoing color to true.
                ## In the next calls, the algorithm looks for vertices compatible with this vertex.
                hypergraph_object.ongoing_color = True
                break
                
        if(hypergraph_object.ongoing_color = True):
            Proper_coloring(hypergraph_object)
            
    else:
        ## This case represents ongoing color = True
        ## The ongoing color is looking for more vertices to repeat itself at.
        
        rep_list = hypergraph_object.repeatable_list()
        ## Repeatable list with respect to ongoing color.
        ## This list will hold all the vertices where the current color may be repeated
        ## It is calculated by checking compatibility with all the vertices that currently hold the same color.
        
        if(len(rep_list) == 0):
            ## No more repetitions possible
            Proper_coloring(hypergraph_object)
            
        
        elif(len(rep_list) == 1):
            ## We just have one vertex where the color may be repeated.
            ## So, directly color that vertex.
            hypergraph_object.assign_color(rep_list[0],current_color)
            Proper_coloring(hypergraph_object)
            
            
        
        elif(len(rep_list) > 1):
            ## Multiple options of potential vertices to choose from.
        
            ## Initialise an array of hypergraph objects to be used for comparison
            comparison_array = []
            
            for vertex in rep_list:
                ## We need to determine coloring which vertex amongst the vertices where repeated coloring
                ## can be made, gives us the best coloring.
                ## Therefore, we try each, perform the further coloring, 
                ## store the colored hypergraph and compare chromatic numbers to make our choice
                
                hypergraph_object.assign_color(vertex,current_color)
            
                leftover_hypergraph = hypergraph_object.uncolored_sub_hypergraph()
                colored_hypergraph = hypergraph_object.colored_sub_hypergraph()
            
                hypergraph_object_new = colored_hypergraph + Proper_coloring(leftover_hypergraph)
                comparison_array.append(hypergraph_object_new)
        
            final_hypergraph = comparison_array[0]
            for hypergraph in comparison_array:
                ##Find hypergraph with least chromatic number
                if(hypergraph.return_chromatic_number() < final_hypergraph.return_chromatic_number()):
                    final_hypergraph = hypergraph
        
            return final_hypergraph
        
      
            
            
            
            
    
            
            
            
            
            
        
            
    
            
                
                
                
                
```
